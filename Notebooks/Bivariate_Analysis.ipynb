## Bivariate Analysis

Is used to examine the relationship between two variables to determine if there is assosiation, correlation or causation between them.

```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# Read data from excel file
df = pd.read_csv('Student_performance_data .csv')

```

## We will conduct a analysis on the relationships of 
 (Gender, Ethnicity, ParentalEducation, StudyTimeWeekly, Absences, Tutoring, ParentalSupport, Extracurricular, Volunteering) with (GPA)

1. Study time weekly does increase GPA but by a small amount.

```python
df_cleaned = df.drop(columns=[
    'StudentID','Age','Gender','Ethnicity','ParentalEducation','Absences','Tutoring','ParentalSupport','Extracurricular','Sports','Music','Volunteering','GradeClass',
])
# Outer Figure Size
plt.figure(figsize=(20, 10))
for i, col in enumerate(df_cleaned.columns): 
 sns.scatterplot(data=df, x='StudyTimeWeekly', y='GPA')
plt.show()
```

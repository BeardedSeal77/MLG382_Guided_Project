## Logistic Regresion Model

```python
df = pd.read_csv('../../Data/Student_performance_non_scaled.csv')
# Feature engineering
def engineer_features(df):
    # Create copies
    df = df.copy()
```

```python
# Grade Class is target veriable
# Get all rows and columns except the last one
X = df.iloc[:, :-1]  
# Get all rows and only the last column
y = df.iloc[:, -1]   

# Train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state=0)
```

```python
# Standardize numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create and train model
logreg = LogisticRegression(max_iter=2000)
logreg.fit(X_train_scaled, y_train)
```

```python
# Test Data
y_pred = logreg.predict(X_test_scaled)
cm_test = confusion_matrix(y_test, y_pred)
print(cm_test)
_ = sns.heatmap(cm_test/np.sum(cm_test), annot=True,cmap='Blues', fmt='.1%')

# Test Accuracy
test_accuracy = accuracy_score(y_test, y_pred)
print("\nTest Accuracy is:", round(test_accuracy * 100,3,), "%" )
```

```python
# Feature importance
result = permutation_importance(logreg, X_test_scaled, y_test, n_repeats=10,random_state=42)

# Create DataFrame to ignore negative values and sort values 
importance_df = pd.DataFrame({'Feature': X.columns,'Importance': result['importances_mean']}).query("Importance > 0").sort_values('Importance', ascending=False)

print(importance_df)
```

```python
# Plot feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importance_df.head(9))
plt.title('Top 9 Feature Importances')
plt.tight_layout()
plt.savefig('../../artifacts/plots/LR_feature_importance.png')

# Test Accuracy
test_accuracy = accuracy_score(y_test, y_pred)
print("\nTest Accuracy is:", round(test_accuracy * 100,3,), "%" )

# Confusion matrix
plt.figure(figsize=(10, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.savefig('../../artifacts/plots/LR_confusion_matrix.png')


# Classification report as PNG
plt.figure(figsize=(10, 6))
plt.text(0.01, 0.99, classification_report(y_test, y_pred))
plt.axis('off')
plt.title('Classification Report')
plt.savefig('../../artifacts/plots/LR_classification_report.png')   
        
```
